```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection


def run_graphs(file): 
    df = pd.read_csv(file)
    df = df.dropna()
    x = df["X"]
    y = df["Y"]
    z = df["Z"]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")

    triangles = ax.plot_trisurf(x,y,z, cmap="viridis", edgecolor="none", alpha=0.8)


    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
    ax.set_zlabel("z axis")

    fig.colorbar(triangles, ax=ax, shrink=0.5)

    ax.view_init(elev=0, azim=0)

    plt.show()


```

```python

run_graphs("./Data/MaySouth.csv")
run_graphs("./Data/AugustSouth.csv")
run_graphs("./Data/MosaicSouth.csv")



```

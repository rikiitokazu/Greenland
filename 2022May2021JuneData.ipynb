# Comparing 2022 May data to 2021 June Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the Excel sheet
file2021 = "./Data/2021JuneData.csv"
file2022 = "./Data/2022MayCrossSection.csv"

df = pd.read_csv(file2021)  


# Extract Distance and Elevation columns
distance = df["Distance"]
elevation = df["Elevation"]

# Create a line plot
plt.plot(distance, elevation, linestyle='-', linewidth=1, color="blue", label="2021 June")


# Creating plot for 2022 Data
df2 = pd.read_csv(file2022)

distance2 = df2["Distance"]
elevation2 = df2["Elevation"]

plt.plot(distance2, elevation2, linestyle="-", linewidth=1, color="red", label="2022 May")

# Set labels and title
plt.xlabel('Distance')
plt.ylabel('Elevation')
plt.title('Difference of May 2022 and June 2021 Elevation ')

plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fileDifference = "./Data/2022May2021JuneDifference.csv"

df = pd.read_csv(fileDifference)
distance = df["Distance"]
elevation = df["Elevation"]

plt.plot(distance, elevation, linestyle="-", linewidth=1, color="black", label="Difference")

# Set labels and title
plt.xlabel('Distance')
plt.ylabel('Elevation')
plt.title('Difference of May 2022 and June 2021 Elevation ')
plt.ylim(-10,10) 
plt.grid(True)
plt.show()
```

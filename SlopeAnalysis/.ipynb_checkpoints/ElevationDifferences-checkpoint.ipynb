Following is for South Lake

```python
# South X values
south_x_values = [976, 976.5, 977, 977.5, 978, 978.5, 979, 979.5, 980]
                  
# August 28, 2022 Y Values
aug_Sy_values = [0.006734, 0.008108, 0.009635, 0.01129, 0.01302, 0.01478, 0.01657, 0.01837, 0.02020]

# May 1, 2022 Values
may_Sy_values = [0.004229, 0.005255, 0.006449, 0.007794, 0.009289, 0.01090, 0.01261, 0.01437, 0.01615]

# Mosaic Values
mosaic_Sy_values = [0.002663, 0.003224, 0.003866, 0.004598, 0.005432, 0.006367, 0.007415, 0.008585, 0.009881]
```

```python
# Creating data to export to excel file

import pandas as pd

north_data = {"North_X": north_x_values, "Aug_NY": aug_Ny_values, "May_NY": may_Ny_values, "Mosaic_NY": mosaic_Ny_values}

south_data = {"South_X": south_x_values, "Aug_SY": aug_Sy_values, "May_SY": may_Sy_values, "Mosaic_SY": mosaic_Sy_values}

north_df = pd.DataFrame(north_data)
north_path = "./Data/north_volumes.csv"

north_df.to_csv(north_path, index=False)


south_df = pd.DataFrame(south_data)
south_path = "./Data/south_volumes.csv"

south_df.to_csv(south_path, index=False)
```

```python
def make_north_graphs(): 
    li = [aug_Ny_values, may_Ny_values, mosaic_Ny_values]
    for dem in li: 
        plt.plot(north_x_values, dem)
        #plt.xlim(963, 980)  # Set x-axis limits
        #plt.ylim(0.0009, 0.03)  # Set y-axis limits        
        plt.xlabel("Z cross sections")
        plt.ylabel("Elevations")
        plt.show()
```

```python
make_north_graphs()
```

```python
def make_south_graphs(): 
    li = [aug_Sy_values, may_Sy_values, mosaic_Sy_values]
    for dem in li: 
        plt.plot(south_x_values, dem)     
        plt.xlabel("Z cross sections")
        plt.ylabel("Elevations")
        plt.show()
```

```python
make_south_graphs()
```

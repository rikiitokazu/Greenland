```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd 


def volume_graphs(file, type_x, type_y):
    df = pd.read_csv(file)
    south_x = df[type_x]
    var_y = df[type_y]
    # Sample x and y values
    x = np.array(south_x)
    y = np.array(var_y)

    # Reshape x to a column vector
    x = x.reshape(-1, 1)

    # Create a linear regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(x, y)

    # Predict y values based on the linear regression model
    y_pred = model.predict(x)

    # Plot the actual and predicted values
    plt.scatter(x, y, label='Actual')
    plt.plot(x, y_pred, color='red', label='Linear Regression', alpha=0.3)
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.title(type_y)
    plt.legend()
    plt.show()

    # Assess linearity using the coefficient of determination (R-squared)
    r_squared = model.score(x, y)
    print(f'R-squared: {r_squared}')
```

```python
# For south lakes

volume_graphs("./Data/south_volumes.csv", "South_X", "May_SY")
volume_graphs("./Data/south_volumes.csv", "South_X", "Aug_SY")
volume_graphs("./Data/south_volumes.csv", "South_X", "Mosaic_SY")
```
